## Extracting results and comparing

There are two approaches to extract results, the first one is out and the second within of OSTRICH-SWMM.

## Option 2:
For this option the .inp and .out files of all the runs to be compared must be stored in the same directory

```python
#Runs the created files
import csv
import datetime as dt
import numpy as np
import os
import swmmtoolbox
import pandas as pd
```

```python
directory= os.getcwd()
# List all files in a directory using scandir()
basepath = os.getcwd()
first=True
names=[]
with os.scandir(basepath) as entries:
    for entry in entries:
        if 'mod' in entry.name:
            if first==True :
                data=pd.read_csv(f"{basepath}\\{entry.name}\model_results.csv", sep=",")
                data['modelo']=entry.name
                first=False
            else :
                data_temp=pd.read_csv(f"{basepath}\\{entry.name}\model_results.csv", sep=",")
                data_temp['modelo']=entry.name
                data=data.append(data_temp)                   

data=data.set_index('modelo')                
if os.path.exists("results.xlsx")==True : os.remove("results.xlsx") #erase the file if exists
data.to_excel("results.xlsx") 

data.head(50)
```
